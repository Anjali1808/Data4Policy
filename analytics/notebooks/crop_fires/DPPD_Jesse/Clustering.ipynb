```python
import pandas as pd
import os
import geopandas as gpd
import geoplot
import geoplot.crs as gcrs
import matplotlib.pyplot as plt
from shapely.geometry import Point
import folium
import rasterio
from sklearn.cluster import KMeans
import numpy as np
```

```python
os.chdir('c:\\Users\\Jesse\\OneDrive\\Documenten\\Master BAOR\\Thesis\\GitHub\\dicra\\analytics\\notebooks\\crop_fires\\DPPD_Jesse')
rwi_data = gpd.read_file('RWI_01-01-2021.geojson')
rwi_data['zonalstat'].iloc[0]['mean']
```

```python
rwi_data['mean_rwi'] = 0

for i in range(0, len(rwi_data)):
    rwi_data['mean_rwi'].iloc[i] = rwi_data['zonalstat'].iloc[i]['mean']
```

```python
X = np.array(rwi_data['mean_rwi'])
X = X.reshape(-1, 1)
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
rwi_data['cluster'] = kmeans.labels_
```

```python
rwi_data.plot(column = 'cluster', legend = True)
plt.show()
```
