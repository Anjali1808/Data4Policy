```python
import pandas as pd
import os
import geopandas as gpd
import geoplot
import geoplot.crs as gcrs
import matplotlib.pyplot as plt
import numpy as np
```

```python
os.chdir('c:\\Users\\Jesse\\OneDrive\\Documenten\\Master BAOR\\Thesis\\GitHub\\dicra\\src\\data_preprocessing\\tsdm\\')
district_boundaries = gpd.read_file('District_Boundary.shp')
district_boundaries = district_boundaries.reset_index()

mandal_boundaries = gpd.read_file('Mandal_Boundary.shp')
mandal_boundaries = mandal_boundaries.reset_index()
```

```python
district_boundaries['coords'] = district_boundaries['geometry'].apply(lambda x: x.representative_point().coords[:])
district_boundaries['coords'] = [coords[0] for coords in district_boundaries['coords']]

district_boundaries.plot(figsize = [20, 16], alpha=0.6, color = 'black')
for idx, row in district_boundaries.iterrows():
    plt.annotate(s=row['Dist_Name'], xy=row['coords'],
                 horizontalalignment='center', fontsize=9)

plt.title('Telangana Districts')
plt.axis('off')
plt.show()
```

```python
mandal_boundaries['coords'] = mandal_boundaries['geometry'].apply(lambda x: x.representative_point().coords[:])
mandal_boundaries['coords'] = [coords[0] for coords in mandal_boundaries['coords']]

mandal_boundaries.plot(figsize = [50, 30], alpha=0.6, color = 'black')
for idx, row in mandal_boundaries.iterrows():
    plt.annotate(s=row['Mandal_Nam'], xy=row['coords'],
                 horizontalalignment='center', fontsize=7)

plt.title('Telangana Mandals')
plt.axis('off')
plt.show()
```
