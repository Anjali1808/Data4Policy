```python
# Import packages
import os
import geopandas as gpd
import numpy as np
import rasterio
import rasterio.mask
```

```python
# Import this file to derive the polygon geometry for Telangana
bound_Telangana = gpd.read_file('C:/Users/mieke/Documents/Msc Thesis/GitHub repositories/dicra/analytics/sandbox/geospatial_internship/datasets/telangana_shapefile.geojson')
bound_Telangana.head()
# Import this file to derive the polygon geometry for Karnataka
bound_Karnataka = gpd.read_file('C:/Users/mieke/Documents/Msc Thesis/Datasets/Shapefiles/karnataka_shapefile.geojson')
bound_Karnataka.head()
```

The following cell is optional.

```python
# Create a Polygon geometry of Telangana
bound_Telangana = bound_Telangana.explode(index_parts = True)
bound_Telangana.head()
# Create a Polygon geometry of Karnataka
bound_Karnataka = bound_Karnataka.explode(index_parts = True)
bound_Karnataka
```

Change the geometry to the correct state of interest.

```python
# For each of the NDVI images, we crop the Geotiff file to Leaf Area Index in Telangana
path = 'C:/Users/mieke/Documents/Msc Thesis/Datasets/Copernicus/NDVI/NDVI_2014_2019/original/'
folder_names = os.listdir(path)
for i in folder_names:
    file_names = os.listdir(path + i)
    for j in file_names:
        NDVI_data = path + i + "/" + j
        with rasterio.open(NDVI_data) as src:
            out_image, out_transform = rasterio.mask.mask(src, bound_Telangana.geometry, crop=True)
            out_meta = src.meta

        out_meta.update({"driver": "GTiff",
                        "height": out_image.shape[1],
                        "width": out_image.shape[2],
                        "transform": out_transform})

        # photometric = 'YCbCr'
        # photometric = 'RGB', compress = 'JPEG'
        # Save the Geotiff file to Leaf Area Index in Telangana
        with rasterio.open('C:/Users/mieke/Documents/Msc Thesis/Datasets/Copernicus/NDVI/NDVI_2014_2019/Telangana/aoi/' + j, "w", **out_meta) as dest:
            dest.write(out_image)
```

Below, we find the exact same approach used for Karnataka.

```python
# For each of the NDVI images, we crop the Geotiff file to Leaf Area Index in Karnataka
path = 'C:/Users/mieke/Documents/Msc Thesis/Datasets/Copernicus/NDVI/NDVI_2014_2019/original/'
folder_names = os.listdir(path)
for i in folder_names:
    file_names = os.listdir(path + i)
    for j in file_names:
        NDVI_data = path + i + "/" + j
        with rasterio.open(NDVI_data) as src:
            out_image, out_transform = rasterio.mask.mask(src, bound_Karnataka.geometry, crop=True)
            out_meta = src.meta

        out_meta.update({"driver": "GTiff",
                        "height": out_image.shape[1],
                        "width": out_image.shape[2],
                        "transform": out_transform})

        # photometric = 'YCbCr'
        # photometric = 'RGB', compress = 'JPEG'
        # Save the Geotiff file to Leaf Area Index in Karnataka
        with rasterio.open('C:/Users/mieke/Documents/Msc Thesis/Datasets/Copernicus/NDVI/NDVI_2014_2019/Karnataka/aoi/' + j, "w", **out_meta) as dest:
            dest.write(out_image)
```
