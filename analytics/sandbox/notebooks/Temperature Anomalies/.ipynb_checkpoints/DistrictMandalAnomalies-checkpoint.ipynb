# Notebook 5: Calculating district and mandal level anomalies 

```python
import os
import numpy as np
import pandas as pd
from osgeo import gdal
import matplotlib.pyplot as plt
# netCDF4 needs to be installed in your environment for this to work
import xarray as xr
import rioxarray as rio
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import seaborn as sns
import geopandas as gpd
import earthpy as et
# Plotting options
sns.set(font_scale=1.3)
sns.set_style("white")
```

```python
district_path = "convert\\districts" #give the paths to the data
mandal_path = "convert\\mandals"
```

```python
district_dataset = [ds for ds in os.listdir(district_path) if ds[-8:] == '.geojson'] #import the data
mandal_dataset = [ds for ds in os.listdir(mandal_path) if ds[-8:] == '.geojson']
```

# District Anomalies

```python
sample =  gpd.read_file(district_path + "/" + district_dataset[870])
```

```python
final_data = []
district_df = pd.DataFrame()

for i in range(0,len(district_dataset)): #for each year
    x = gpd.read_file(district_path + "/" + district_dataset[i])
    district_df["District"] = x["Dist_Name"]
    d_data =  []
    for j in range(0, len(x)): #get one district at a time 
        d_data.append(x["zonalstat"][j]["mean"])
    final_data.append(d_data)
    #print(x["Dist_Name"][0])
    #print(x["zonalstat"][0]["max"])
```

```python
district_df = pd.DataFrame(final_data) #store the data of all the districts into a single dataframe.
district_df = district_df.T
```

```python
district_df["District"] = sample["Dist_Name"] #get names and geometry from a single object.
district_df["geometry"] = sample["geometry"]
district_df.head(2)
```

```python
#Plotting and understanding trends observed in districts of telangana.
fig, axes = plt.subplots(11,3,figsize=(20,16))
x = 0
for i in range(0,11):
    for j in range(0,3):
        if x > 871:
            break
        data = district_df.iloc[x]
        axes[i][j].set_title("Temperature anomalies for " + str(data["District"]))
        axes[i][j].plot(data[0:871])
        x = x+1
fig.tight_layout()
plt.show()
```

```python
arr = []

for i in range(0, len(sample)):
    arr.append(sample["zonalstat"][i]["mean"])
sample["mean"] = arr #arr has all the mean values of zonalstat-temperature anomalies 
```

```python
#Visualizing the temperature anomalies for districts
plt.figsize=(9,9)
#ax.set_aspect('equal')
sample.plot(column="mean", cmap='RdYlGn_r', legend=True)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('District level Temperature Anomalies in Telangana (July 2022)',fontsize=15)
#ax.set_axis_off()
```

# Mandal Anomalies

```python
sample =  gpd.read_file(mandal_path + "/" + mandal_dataset[870])
```

```python
mandal_df = pd.DataFrame()
arr = []

for i in range(0, len(sample)):
    arr.append(sample["zonalstat"][i]["mean"])
```

```python
sample["mean"] = np.array(arr) #get mean temperature anomaly values from the data.
```

```python
final_data = []

for i in range(0,len(mandal_dataset)): #for each year
    x = gpd.read_file(mandal_path + "/" + mandal_dataset[i])
    d_data =  []
    for j in range(0, len(x)): #get one district at a time 
        d_data.append(x["zonalstat"][j]["mean"])
    final_data.append(d_data)
```

```python
mandal_df = pd.DataFrame(final_data)
```
