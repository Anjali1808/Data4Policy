# Notebook 4: Preparation of data to get temperature anomalies for district and mandals of Telangana
We prepare it by creating several tif images of the telagana data

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# netCDF4 needs to be installed in your environment for this to work
import xarray as xr
import rioxarray as rio
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import seaborn as sns
import geopandas as gpd
import earthpy as et
# Plotting options
sns.set(font_scale=1.3)
sns.set_style("white")
```

```python
xr.set_options(keep_attrs= False)

telangana_anomaly_path = "telanganaanomaly.nc" #import telangana anomalies file
telangana_anomaly = xr.open_dataset(telangana_anomaly_path)
telangana_anomaly = telangana_anomaly.__xarray_dataarray_variable__
tel = telangana_anomaly
```

```python
tel.coords['longitude'] = (tel.coords['longitude'] + 180) % 360 - 180 #get coordinates
tel = tel.sortby(tel.longitude)
tel = tel.rio.set_spatial_dims('longitude', 'latitude')
```

```python
tel.rio.write_crs("epsg:4326", inplace=True)
```

```python
#tel = tel.drop(tel.attrs['grid_mapping'])
del tel.attrs['grid_mapping']
tel.rio.to_raster(raster_path="telanganaanomaly.tif")
```

```python
#get all the data into a folder to 
size = len(tel)
for i in range (0, size):
    tel[i].rio.to_raster(raster_path="convert//telanomaly//telanganaanomaly"+ str(i) + ".tif")
```
