```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
adm_name = gpd.read_file('data/TSDM/District_Boundary.shp')
df = pd.read_csv('data/cropfire_dist.csv', index_col=False)
adm_name = adm_name[adm_name.Dist_Name != 'Hyderabad']      #Found that this district doesnt have any crop fire points in data preprocessing part
```

```python
dfdist = df.pivot_table('fireCount', 'Date', 'place')
```

```python
dfdist.plot(figsize=(20, 12))
```

```python
df = pd.read_csv('data/cropfire_dist.csv', index_col=False)
dfcount = df.pivot_table('fireCount', 'place')
```

```python
adm_name = adm_name.sort_values(by='Dist_Name')
```

```python
adm_name['fireCount']=list(dfcount['fireCount'])
```
